Total gross revenue by year(Domestic and foreign)

```python
total_gross_revenue_by_year=Bom_movie_cleaned.groupby('year')[['domestic_gross', 'foreign_gross']].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data = total_gross_revenue_by_year, x='year', y='domestic_gross', label='Domestic Gross')
sns.lineplot(data = total_gross_revenue_by_year, x='year', y='foreign_gross', label='Foreign Gross')

plt.title('Total Gross Revenue Yearly Trend')
plt.xlabel('Year')
plt.ylabel('Revenue (in billions)')
plt.legend()
plt.show()
```

The graph above shows that over the years, foreign revenues from movie production have surpassed domestic revenues.

```python


# Explode genres
df = Basics_Ratings.copy()
df['genres'] = df['genres'].str.split(',')
df = df.explode('genres')

# Get top 15 genres by count
top_15_genres = (
    df['genres']
    .value_counts()
    .head(15)
    .index
)

# Filter for top 15 genres
df_top = df[df['genres'].isin(top_15_genres)]

# Average rating by year and genre
genre_year_avg = (
    df_top
    .groupby(['start_year', 'genres'])['averagerating']
    .mean()
    .reset_index()
)

# Plot
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=genre_year_avg,
    x='start_year',
    y='averagerating',
    hue='genres'
)

plt.title('Average Rating Trends by Year (Top 15 Genres)')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))

ax = sns.lineplot(
    data=genre_year_avg,
    x='start_year',
    y='averagerating',
    hue='genres'
)

plt.title('Average Rating Trends by Year (Top 15 Genres)')
plt.xlabel('Year')
plt.ylabel('Average Rating')

# Get legend handles & labels in line order
handles, labels = ax.get_legend_handles_labels()

# Re-apply legend in the same order as plotted
ax.legend(
    handles=handles,
    labels=labels,
    title='Genre',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))

ax = sns.lineplot(
    data=genre_year_avg,
    x='start_year',
    y='averagerating',
    hue='genres',
    legend=False  # remove legend
)

# Add text labels at the end of each line
for genre in genre_year_avg['genres'].unique():
    genre_data = genre_year_avg[genre_year_avg['genres'] == genre]

    # Get last point
    last_year = genre_data['start_year'].max()
    last_rating = genre_data.loc[
        genre_data['start_year'] == last_year, 'averagerating'
    ].values[0]

    ax.text(
        last_year + 0.2,      # small shift to the right
        last_rating,
        genre,
        fontsize=9,
        va='center'
    )

plt.title('Average Rating Trends by Year (Top 15 Genres)')
plt.xlabel('Year')
plt.ylabel('Average Rating')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 8))

sns.boxplot(
    data=df_top,
    x='genres',
    y='runtime_minutes'
)

plt.title('Distribution of Movie Runtime by Genre (Top 15 Genres)')
plt.xlabel('Genre')
plt.ylabel('Runtime (minutes)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
df = Basics_Ratings.copy()
df = df.dropna(subset=['runtime_minutes', 'genres'])
df['genres'] = df['genres'].str.split(',')
df = df.explode('genres')

# Top 15 genres
top_15_genres = df['genres'].value_counts().head(15).index
df_top = df[df['genres'].isin(top_15_genres)]
bins = np.arange(0, df_top['runtime_minutes'].max() + 30, 30)  # bin width = 30 min
counts_df = pd.DataFrame()

for genre in top_15_genres:
    genre_runtimes = df_top[df_top['genres'] == genre]['runtime_minutes']
    counts, _ = np.histogram(genre_runtimes, bins=bins)
    counts_df[genre] = counts

counts_df['bin_start'] = bins[:-1]
plt.figure(figsize=(16, 8))

# Plot each genre as a stack
bottom = np.zeros(len(counts_df))
colors = plt.cm.tab20.colors  # nice colormap

for i, genre in enumerate(top_15_genres):
    plt.bar(
        counts_df['bin_start'],
        counts_df[genre],
        bottom=bottom,
        width=30,
        color=colors[i % len(colors)],
        label=genre,
        alpha=0.8,
        edgecolor='k'
    )
    bottom += counts_df[genre]
plt.plot(
    counts_df['bin_start'],
    bottom,
    color='black',
    linewidth=2,
    label='Total Movies')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movies by Runtime: Top 15 Genres')
plt.xticks(bins)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
